```python
plt.scatter(df.index, df['sepal_length'])
plt.scatter(df.index, df['sepal_width'])
plt.scatter(df.index, df['petal_length'])
plt.scatter(df.index, df['petal_width'])
plt.show()
```

```python
sns.scatterplot(x=df.index, y=df['sepal_length'], hue=df['species'])
```

# Line Plot

```python
plt.figure(figsize=(6,6))
plt.xlabel('index', fontsize=20)
plt.ylabel('sepal_length', fontsize=20)
plt.plot(df.index, df['sepal_length'], markevery=1, marker='d')
for name, group in df.groupby('species'):
    plt.plot(group.index, group['sepal_length'], label=name, markevery=1, marker='d')
plt.legend()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.set(font_scale=1.5)
```

```python
fig=sns.lineplot(x=df.index, y=df['sepal_length'],markevery=1,marker='d',data=df,hue=df['species'])
fig.set(xlabel='index')
```

# Strip Plot

```python
sns.stripplot(x=df['species'], y=df['sepal_width'])
```

# Histogram

```python
plt.hist(df['sepal_width'])
```

# Box Plot

```python
dfM=df.drop('species', axis=1)  #removing the columns with categorical variables
sns.set(rc={'figure.figsize':(9,9)})
sns.boxplot (x="variable", y="value", data=pd.melt(dfM))
```

# Violin Plot

```python
sns.set(rc={'figure.figsize':(9,9)})
sns.violinplot(x=df['species'], y=df['sepal_width'], data=df)
```
